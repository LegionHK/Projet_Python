**Online News Popularity Data Set**

This dataset summarizes a heterogeneous set of features about articles published by Mashable in a period of two years. The goal is to predict the number of shares in social networks (popularity).


https://archive.ics.uci.edu/ml/datasets/Online+News+Popularity

**Attribute Information:**

Number of Attributes: 61 (58 predictive attributes, 2 non-predictive, 1 goal field)

Attribute Information:
0. url: URL of the article (non-predictive)
1. timedelta: Days between the article publication and the dataset acquisition (non-predictive)
2. n_tokens_title: Number of words in the title
3. n_tokens_content: Number of words in the content
4. n_unique_tokens: Rate of unique words in the content
5. n_non_stop_words: Rate of non-stop words in the content
6. n_non_stop_unique_tokens: Rate of unique non-stop words in the content
7. num_hrefs: Number of links
8. num_self_hrefs: Number of links to other articles published by Mashable
9. num_imgs: Number of images
10. num_videos: Number of videos
11. average_token_length: Average length of the words in the content
12. num_keywords: Number of keywords in the metadata
13. data_channel_is_lifestyle: Is data channel 'Lifestyle'?
14. data_channel_is_entertainment: Is data channel 'Entertainment'?
15. data_channel_is_bus: Is data channel 'Business'?
16. data_channel_is_socmed: Is data channel 'Social Media'?
17. data_channel_is_tech: Is data channel 'Tech'?
18. data_channel_is_world: Is data channel 'World'?
19. kw_min_min: Worst keyword (min. shares)
20. kw_max_min: Worst keyword (max. shares)
21. kw_avg_min: Worst keyword (avg. shares)
22. kw_min_max: Best keyword (min. shares)
23. kw_max_max: Best keyword (max. shares)
24. kw_avg_max: Best keyword (avg. shares)
25. kw_min_avg: Avg. keyword (min. shares)
26. kw_max_avg: Avg. keyword (max. shares)
27. kw_avg_avg: Avg. keyword (avg. shares)
28. self_reference_min_shares: Min. shares of referenced articles in Mashable
29. self_reference_max_shares: Max. shares of referenced articles in Mashable
30. self_reference_avg_sharess: Avg. shares of referenced articles in Mashable
31. weekday_is_monday: Was the article published on a Monday?
32. weekday_is_tuesday: Was the article published on a Tuesday?
33. weekday_is_wednesday: Was the article published on a Wednesday?
34. weekday_is_thursday: Was the article published on a Thursday?
35. weekday_is_friday: Was the article published on a Friday?
36. weekday_is_saturday: Was the article published on a Saturday?
37. weekday_is_sunday: Was the article published on a Sunday?
38. is_weekend: Was the article published on the weekend?
39. LDA_00: Closeness to LDA topic 0
40. LDA_01: Closeness to LDA topic 1
41. LDA_02: Closeness to LDA topic 2
42. LDA_03: Closeness to LDA topic 3
43. LDA_04: Closeness to LDA topic 4
44. global_subjectivity: Text subjectivity
45. global_sentiment_polarity: Text sentiment polarity
46. global_rate_positive_words: Rate of positive words in the content
47. global_rate_negative_words: Rate of negative words in the content
48. rate_positive_words: Rate of positive words among non-neutral tokens
49. rate_negative_words: Rate of negative words among non-neutral tokens
50. avg_positive_polarity: Avg. polarity of positive words
51. min_positive_polarity: Min. polarity of positive words
52. max_positive_polarity: Max. polarity of positive words
53. avg_negative_polarity: Avg. polarity of negative words
54. min_negative_polarity: Min. polarity of negative words
55. max_negative_polarity: Max. polarity of negative words
56. title_subjectivity: Title subjectivity
57. title_sentiment_polarity: Title polarity
58. abs_title_subjectivity: Absolute subjectivity level
59. abs_title_sentiment_polarity: Absolute polarity level
60. shares: Number of shares (target)

### Outliers values

```python
data['shares'].plot(kind = 'box')
```

```python
Q3 = np.percentile(data['shares'], 75)

Q1 = np.percentile(data['shares'], 25)

IQR = Q3 - Q1

Outliers = Q3+1.5*IQR

data = data[data['shares']<Outliers]

data['shares'].plot(kind = 'box')
```

```python
data.shares.describe()
```

```python
popular = [1 if value > data['shares'].median() else 0 for value in data["shares"]]
```

```python
data['popular'] = popular
```

## Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.feature_selection import RFECV

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import mean_squared_error
```

we remove the two 'non-predictive' columns

```python
X = data.drop(columns = ['url','timedelta','shares','popular'])
y = data['shares']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

### Feature selection

```python
model_lr = LinearRegression(normalize=True)
model_rfr = RandomForestRegressor(n_jobs=-1, max_samples= 1000)

selector_lr = RFECV(model_lr, min_features_to_select = 10 , cv = 5)
selector_lr.fit(X,y)

selector_rfr = RFECV(model_rfr, min_features_to_select = 10, cv = 5)
selector_rfr.fit(X,y)
```

```python
selector_lr.get_support()
```

```python
selector_rfr.get_support()
```

### Data Modeling

```python
X_lr = X[X.columns[selector_lr.get_support()]]
X_train, X_test, y_train, y_test = train_test_split(X_lr,y,test_size=0.2)
```

```python
lr = LinearRegression(normalize=True)
lr.fit(X_train,y_train)

y_predict_lr = lr.predict(X_test)

print( "MSE : ",mean_squared_error(y_true=y_test,y_pred=y_predict_lr))
print("score : ",lr.score(X_test,y_test))
```

```python
X_rfr = X[X.columns[selector_rfr.get_support()]]
X_train, X_test, y_train, y_test = train_test_split(X_rfr,y,test_size=0.2)
```

```python
rfr = RandomForestRegressor(n_jobs=-1, max_samples= 1000, max_features=18)
rfr.fit(X_train,y_train)

y_predict_rfr = rfr.predict(X_test)

print( "MSE : ",mean_squared_error(y_true=y_test,y_pred=y_predict_rfr))
print("score : ",rfr.score(X_test,y_test))
```

As we don't obtain any good result in regression, we can try with a **classification objective.** 

Now we will try to **determine if an article is popular** instead of the number of shares. 

## Classification

However, we can ask ourself, what is a popular article ?

For us an article is popular if its **number of shares** is at least **above the median**. 

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.feature_selection import RFECV

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

we remove the two 'non-predictive' columns

```python
X = data.drop(columns = ['url','timedelta','shares','popular'])
y = data['popular']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

### Feature Selection

```python
model_rfc = RandomForestClassifier(n_jobs=-1,max_samples=2000)
selector_rfc = RFECV(model_rfc, n_jobs=-1, min_features_to_select=10, cv=5)
selector_rfc.fit(X,y)
```

```python
model_sgdc = SGDClassifier(n_jobs=-1)
selector_sgdc = RFECV(model_sgdc, n_jobs=-1, min_features_to_select=10, cv=5)
selector_sgdc.fit(X,y)
```

```python
selector_rfc.get_support()
```

```python
selector_rfc.ranking_
```

```python
selector_sgdc.get_support()
```

### Data Modeling

```python
X_sgdc = X[X.columns[selector_sgdc.get_support()]]
X_train, X_test, y_train, y_test = train_test_split(X_sgdc,y,test_size=0.2)
```

```python
sgdc = SGDClassifier(n_jobs=-1)
sgdc.fit(X_train,y_train)

y_predict_sgdc = sgdc.predict(X_test)

print( "MSE : ",mean_squared_error(y_true=y_test,y_pred=y_predict_sgdc))
print("score : ",sgdc.score(X_test,y_test))
```

```python
X_rfc = X[X.columns[selector_rfc.get_support()]]
X_train, X_test, y_train, y_test = train_test_split(X_rfc,y,test_size=0.2)
```

```python
rfc = RandomForestClassifier(n_jobs=-1 ,  max_samples= 8500,max_features=9, max_depth=12)
rfc.fit(X_train,y_train)

y_predict_rfc = rfc.predict(X_test)

print( "MSE : ",mean_squared_error(y_true=y_test,y_pred=y_predict_rfc))
print("score : ",rfc.score(X_test,y_test))
```

### Hyperparameter Tuning 

#### GridSearch CV 

```python
param_dict = {
    'max_samples' : [3000,6000,8000],
    'max_features' : [5,8,9,18,19],
    'max_depth' : [12,15,18,21]
}

grid = GridSearchCV(RandomForestClassifier(), param_grid=param_dict, cv = 5, n_jobs=-1, verbose = 3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
rfc = grid.best_estimator_

rfc.fit(X_train,y_train)

y_predict_rfc = rfc.predict(X_test)

print( "MSE : ",mean_squared_error(y_true=y_test,y_pred=y_predict_rfc))
print("score : ",rfc.score(X_test,y_test))
```

#### Learning Curve

we are going to draw **3 learning curves** from the following parameters:
1. max_samples : The number of samples to draw from X to train each base estimator.
2. max_features : The number of features to consider when looking for the best split.
3. max_depth : The maximum depth of the tree.

```python
accuracy_max_samples = []

for samples in range(1,9000,10):
    rfc = RandomForestClassifier(n_jobs=-1, max_samples= samples,max_features=29, max_depth=12)
    rfc.fit(X_train,y_train)
    Z = rfc.predict(X_test)
    accuracy_max_samples.append(rfc.score(X_test,y_test))
```

```python
accuracy_max_features = []

for samples in range(1,50):
    rfc = RandomForestClassifier(n_jobs=-1, max_samples= 6000,max_features=samples, max_depth=12)
    rfc.fit(X_train,y_train)
    accuracy_max_features.append(rfc.score(X_test,y_test))
```

```python
accuracy_max_depth = []

for samples in range(1,50):
    rfc = RandomForestClassifier(n_jobs=-1, max_samples= 6000,max_features=29, max_depth=samples)
    rfc.fit(X_train,y_train)
    accuracy_max_depth.append(rfc.score(X_test,y_test))
```

```python
import pickle as pk

outfile = open("accuracy_max_samples",'wb')
pk.dump(accuracy_max_samples ,outfile)
outfile.close()

outfile = open("accuracy_max_features",'wb')
pk.dump(accuracy_max_features ,outfile)
outfile.close()

outfile = open("accuracy_max_depth",'wb')
pk.dump(accuracy_max_depth ,outfile)
outfile.close()
```

```python
fig, ax = plt.subplots() 
plt.plot([i for i in range(1,9000,10)], accuracy_max_samples)
plt.title("Accuracy in fonction of the number of samples")
plt.xlabel("Number of samples")
plt.ylabel("Accuracy")
fig.set_size_inches(15, 6, forward=True)
plt.show()
```

```python
fig, ax = plt.subplots() 
plt.plot([i for i in range(1,50)], accuracy_max_features)
plt.title("Accuracy in fonction of the number of features")
plt.xlabel("Number of features")
plt.ylabel("Accuracy")
fig.set_size_inches(15, 6, forward=True)
plt.show()
```

```python
fig, ax = plt.subplots() 
plt.plot([i for i in range(1,50)], accuracy_max_depth)
plt.title("Accuracy in fonction of the depth")
plt.xlabel("Depth")
plt.ylabel("Accuracy")
fig.set_size_inches(15, 6, forward=True)
plt.show()
```

```python
col = X.columns[selector_rfc.get_support()]
    
y = rfc.feature_importances_

fig, ax = plt.subplots() 
width = 0.5 
ind = np.arange(len(y))
ax.barh(ind, y, color="green")
ax.set_yticks(ind+width/10)
ax.set_yticklabels(col)
plt.title("Feature importance in RandomForest Classifier")
plt.xlabel("Relative importance")
plt.ylabel("feature") 
plt.figure(figsize=(5,5))
fig.set_size_inches(6.5, 15, forward=True)
plt.savefig("Feature importance in RandomForest Classifier.png")
```
